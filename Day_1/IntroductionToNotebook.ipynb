```python
# import the python modules we will use
import matplotlib.pyplot as plt  # major plotting module
import numpy as np # major matrix module
```

```python
# Construct a vector
my_vector = np.arange(0, 100, 5) # remember python indexing etc starts at 0
my_vector
```

```python
# Do some plotting
plt.plot(my_vector, np.exp(1/(5+my_vector)), 'r*-'); # note np.exp, we are using exp from the numpy module
```

```python
# Nicer plotting
fig, axs = plt.subplots(1, 2)  # set up the figure and a pair of axes
axs[0].plot(my_vector, my_vector)
axs[0].set_title('linear')
axs[1].plot(my_vector, my_vector**2)
axs[1].set_title('quadratic');
```
